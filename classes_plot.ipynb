```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import utils 
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
data = utils.loadData("./models/augmented_data")
```

```python
new =  []
for d in data:
    new.append(np.reshape(d, -1))
data = np.array(new)
```

```python
print(data.shape)
```

```python
labels = utils.loadData("./models/augmented_labels")
```

```python

pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)
kmeans = KMeans(n_clusters=36) 
clusters = kmeans.fit_predict(data_pca)
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=clusters)
plt.show()

```
